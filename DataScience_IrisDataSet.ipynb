Up to now amara potita pair kore dekhalam, jodi 10/12 column thake or aro besi thake eta bar bar pair kore dekha is boring & difficult, time consuming. We can solve this problem by using "pairplot" as shown below. here ID not feature to drop it (iris.drop); axis = 1 ID deleted by column (as we want here) axis = 0, Id deleted by row. You can read the opening in new tab. amader numerical feature chhilo 4 ta sobar sather sobar plot kore eta 4X4 = 16 image show korchhe

```python
sns.pairplot(iris.drop('Id', axis = 1), hue='Species', height = 9)
```
